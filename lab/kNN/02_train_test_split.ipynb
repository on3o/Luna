# 数据切分
机器学习需要把数据分为训练集和数据集，因为我们训练出来的模型并不能“确保”有效性或适用性，所以必须在同一个已知数据集中选出一些数据集进行训练，另外一些数据集进行测试。训练数据集和测试数据集都是确定的一一的对应关系，所以我们可以用来进行模型有效性的判断或评分。

## 让我们看一下sklearn中是怎么实现的
1. 引入iris数据集

```
from sklearn.datasets import load_iris
raw_data = load_iris()
raw_data
```

```
X = raw_data.data
y = raw_data.target
```

2. 使用sklearn进行数据集划分

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
print("验证sklearn中训练数据集的比例为：",112/len(X))
```

3. 数据切分完成

数据验证：

- sklean.kNN

```
from sklearn.neighbors import KNeighborsClassifier as sk_kNNClassifier

sk_knn = sk_kNNClassifier()
sk_knn.fit(X_train,y_train)
sk_knn.score(X_test,y_test)
```

- 使用自己的类验证

```
from swallower.neighbors import KNeighborsClassifier as sw_kNNClassifier

sw_knn = sw_kNNClassifier()
sw_knn.fit(X_train,y_train)
sw_knn.score(X_test,y_test)
```
